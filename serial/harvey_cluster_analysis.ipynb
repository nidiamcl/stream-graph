```python
def getClustersInfo(in_path):
    
    all_files = [f for f in listdir(in_path) if isfile(join(in_path, f))]

    f_dicts = []
    m_dicts = []
    for f in all_files:
        print(f)
        graph = nx.read_gpickle(in_path + f)
        fps, fmap = findClusters(graph, 0.3)
        print('clusters found: ' + str(len(fmap)))

        merged_fps, merged_fmap = mergeFingerprints(fps, fmap, 0.05)
        print('clusters merged: ' + str(len(fmap)-len(merged_fmap)))
        print('remaining clusters: ' + str(len(merged_fmap)))
        print('')
        
        f_dict = {str(k): len(v) for k, v in fmap.items()}
        f_dicts.append(f_dict)
        
        m_dict = {k: len(v) for k, v in merged_fmap.items()}
        m_dicts.append(m_dict)
        
    return f_dicts, m_dicts
```

```python
def showClusters(in_path):
    ''' plots clusters for all nx network objects stored in pkl files'''
    
    all_files = [f for f in listdir(in_path) if isfile(join(in_path, f))]
    
    for f in all_files: 
        fn = f.split('-')
#         print(fn)
        fname = fn[1] + '-' + fn[2].split(' ')[0]
#         print(fname)
        print('network: ' + fname)

        graph = nx.read_gpickle(in_path + f)
        fps, fmap = findClusters(graph, 0.3)
        print('clusters found: ' + str(len(fmap)))

        merged_fps, merged_fmap = mergeFingerprints(fps, fmap, 0.05)
        print('clusters merged: ' + str(len(fmap)-len(merged_fmap)))
        print('remaining clusters: ' + str(len(merged_fmap)))
        
        import matplotlib.pyplot as plt
        plt.rcParams["figure.figsize"] = (20, 20)
              
        fdict = {str(k): len(v) for k, v in fmap.items()}
        name1 = 'Graph clusters before merging fingerprints: ' + fname
        m, x, y = drawClusters(graph, fmap, 'rosybrown', name1, fname)

        import matplotlib.pyplot as plt
        plt.rcParams["figure.figsize"] = (20, 20)

        m_dict = {k: len(v) for k, v in merged_fmap.items()}
        name2 = 'Graph clusters after merging fingerprints: ' + fname
        m, x, y = drawClusters(graph, merged_fmap, 'black', name2 , fname+'_merged')
```

```python
run_info_file = '../../graph_clustering/data/twitter_hurricanes/harvey/graphs_info'
```

```python
with open(run_info_file) as f:
    dates = f.readlines()
    dates = [x.strip() for x in dates]
    dates = [line.split(' ') for line in dates]
```

```python
days = []
nodes = []
edges = []
python_time = []
spark_time = []

for item in dates:
    day = item[0]
    days.append(day)
    num_nodes = item[1]
    nodes.append(num_nodes)
    num_edges = item[2]
    edges.append(num_edges)
    pytime = item[3]
    python_time.append(pytime)
    sptime = item[4]
    spark_time.append(sptime)
```

```python
import pandas as pd
df = pd.DataFrame(list(zip(days, nodes, edges, python_time, spark_time)), columns =['day', 'nodes', 'edges', 'python_time', 'spark_time']) 
```

```python
(lambda n: n*(n-1)/2)
```

```python
df['nodes'] = df['nodes'].astype('int64', copy=False)
df['edges'] = df['edges'].astype('int64', copy=False)
df['python_time'] = df['python_time'].astype('float', copy=False)
df['spark_time'] = df['spark_time'].astype('float', copy=False)
df['max_edges'] = df['nodes'].apply(lambda n: (n*(n-1))/2).astype('int64', copy=False)
df
```
